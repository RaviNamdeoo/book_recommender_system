```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
books = pd.read_csv('data\BX-Books.csv', sep=';', on_bad_lines='skip', encoding='latin-1', usecols=['ISBN','Book-Title','Book-Author','Year-Of-Publication','Publisher','Image-URL-L'])
books.head()
```

```python
books.columns
```

```python
books.rename(columns = {
    'Book-Title':'title',
    'Book-Author':'author',
    'Year-Of-Publication':'year',
    'Image-URL-L':'img_url'
}, inplace=True)
```

```python
users = pd.read_csv('data\BX-Users.csv', sep=';', on_bad_lines='skip', encoding='latin-1')
users.head()
```

```python
ratings = pd.read_csv('data\BX-Book-Ratings.csv', sep=';', on_bad_lines='skip', encoding='latin-1')
ratings.head()
```

```python
print(books.shape)
print(users.shape)
print(ratings.shape)
```

```python
ratings.rename(columns={
    'User-ID':'user_id',
    'Book-Rating':'rating'
}, inplace=True)
```

```python
ratings.head()
```

```python
users.rename(columns={
    'User-ID':'user_id',
    'Location':'location',
    'Age':'age'
},inplace=True)
```

```python
ratings['user_id'].value_counts()
```

```python
ratings['user_id'].nunique()
```

```python
ratings.shape
```

```python
# Users who have rated at least 200 books (1200)

user_counts = ratings['user_id'].value_counts()
filtered_users = user_counts[user_counts >= 200].index
ratings = ratings[ratings['user_id'].isin(filtered_users)]
ratings
```

```python
books.shape
```

```python
books.head(3)
```

```python
# Merging the books and ratings dataframe with the common column 'ISBN' so we get books with more than 200 ratings
ratings_with_books = ratings.merge(books, on='ISBN')
ratings_with_books.shape
```

```python
ratings_with_books.head(3)
```

```python
ratings_with_books['title'].value_counts()
```

```python
# Counts how many ratings each book has received and returns it as a clean DataFrame with title and rating columns
num_ratings = ratings_with_books.groupby('title')['rating'].count().reset_index()
num_ratings.head(3)
```

```python
num_ratings.rename(columns={'rating':'no_of_rating'},inplace=True)
num_ratings.head(3)
```

```python
# merging the ratings_with_books and num_ratings dataframes
final_rating = ratings_with_books.merge(num_ratings, on='title')
final_rating.head(3)
```

```python
final_rating.shape
```

```python
final_rating = final_rating[final_rating['no_of_rating'] >= 50]
final_rating.shape
```

```python
final_rating.sample(7)
```

```python
print(final_rating.duplicated().sum())
print(final_rating.isna().sum())
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.shape
```

```python
final_rating.info()
```

```python
book_pivot = final_rating.pivot_table(index='title',columns='user_id',values='rating')
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
book_sparse
```

# Modeling

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')

model.fit(book_sparse)
```
